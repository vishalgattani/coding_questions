# Climbing Stairs

You are climbing a staircase. It takes n steps to reach the top.

Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

 

Example 1:

Input: n = 2
Output: 2
Explanation: There are two ways to climb to the top.
1. 1 step + 1 step
2. 2 steps
Example 2:

Input: n = 3
Output: 3
Explanation: There are three ways to climb to the top.
1. 1 step + 1 step + 1 step
2. 1 step + 2 steps
3. 2 steps + 1 step
 



```python
def climbStairsRecursive(n: int) -> int:
    # resursive
    def climb(i: int) -> int:
        if i <= 1:
            return 1
        return climb(i - 1) + climb(i - 2)
    
    def climb(i: int, k: int) -> int:
        if i <= 1:
            return 1
        if k == 1:
            return i
        return climb(i - 1, k - 1) + climb(i - 2, k)
    
    return climb(n)

def climbStairsDP(n: int, k: int = 2) -> int:
    return climb(n,k)

def climb(n,atmost_k):
    if n <=1:
        return n
    dp = [0]*(n+1)
    dp[0] = 1
    dp[1] = 1
    for i in range(2,n+1):
        for j in range(1,min(i+1,atmost_k+1)): 
            # all possible steps (from i to k) that could have led to the current stair
            # use min(i + 1, k + 1) to not consider more steps than the current stair number
            dp[i] = dp[i] + dp[i-j] # valid step
    return dp[n]

def ClimbStairsSinglePass(n: int) -> int:
    # single pass
    if n <= 1:
        return n
    prev, curr = 1, 1
    for i in range(2,n+1):
        prev, curr = curr, prev + curr
    return curr
    

print(climbStairsDP(3))
# print(climbStairsRecursive(3))
print(ClimbStairsSinglePass(3))
```

# Coin Change

You are given an integer array coins representing coins of different denominations and an integer amount representing a total amount of money.

Return the fewest number of coins that you need to make up that amount. If that amount of money cannot be made up by any combination of the coins, return -1.

You may assume that you have an infinite number of each kind of coin.

 

Example 1:

Input: coins = [1,2,5], amount = 11
Output: 3
Explanation: 11 = 5 + 5 + 1
Example 2:

Input: coins = [2], amount = 3
Output: -1
Example 3:

Input: coins = [1], amount = 0
Output: 0
 

```python
def coinChange(coins, amount,find_min=True):
    if amount == 0:
        return 0
    elif amount < 0:
        return -1
    else:  
        dp = [-1]*(amount+1) # val is number of ways to make change for the index amount
        dp[0] = 0 # 0 amount with 0 coins
        for i in range(1,amount+1):
            for coin in coins:
                if coin <= i and dp[i-coin]!=-1: # coin value less than amount
                    if dp[i] == -1: # valid coin
                        dp[i] = dp[i-coin]+1
                    # else:
                    if find_min:
                        dp[i] = min(dp[i],dp[i-coin]+1)
                    else:
                        dp[i] = max(dp[i],dp[i-coin]+1)
        return dp[amount]
    
print(coinChange(coins=[1,2,5], amount=11))
print(coinChange(coins=[1,2,5], amount=11,find_min=False))
```

# House Robber Noob

You are a professional robber planning to rob houses along a street. Each house has a certain amount of money stashed, the only constraint stopping you from robbing each of them is that adjacent houses have security systems connected and it will automatically contact the police if two adjacent houses were broken into on the same night.

Given an integer array nums representing the amount of money of each house, return the maximum amount of money you can rob tonight without alerting the police.

Example 1:

Input: nums = [1,2,3,1]
Output: 4
Explanation: Rob house 1 (money = 1) and then rob house 3 (money = 3).
Total amount you can rob = 1 + 3 = 4.
Example 2:

Input: nums = [2,7,9,3,1]
Output: 12
Explanation: Rob house 1 (money = 2), rob house 3 (money = 9) and rob house 5 (money = 1).
Total amount you can rob = 2 + 9 + 1 = 12.

# House Robber Pro

You are a professional robber planning to rob houses along a street. Each house has a certain amount of money stashed. All houses at this place are arranged in a circle. That means the first house is the neighbor of the last one. Meanwhile, adjacent houses have a security system connected, and it will automatically contact the police if two adjacent houses were broken into on the same night.

Given an integer array nums representing the amount of money of each house, return the maximum amount of money you can rob tonight without alerting the police.

 

Example 1:

Input: nums = [2,3,2]
Output: 3
Explanation: You cannot rob house 1 (money = 2) and then rob house 3 (money = 2), because they are adjacent houses.
Example 2:

Input: nums = [1,2,3,1]
Output: 4
Explanation: Rob house 1 (money = 1) and then rob house 3 (money = 3).
Total amount you can rob = 1 + 3 = 4.
Example 3:

Input: nums = [1,2,3]
Output: 3


```python
def noob_rob(nums):
    if not nums:
            return 0
    if len(nums) == 1:
        return nums[0]
    n = len(nums)
    dp = [0] * n
    dp[0] = nums[0]
    dp[1] = max(nums[0], nums[1])
    for i in range(2,n):
        dp[i] = max(dp[i-1],dp[i-2]+nums[i])
    return dp[-1]

print(noob_rob([1,2,3,4]))

def professional_rob(nums):
    if not nums:
        return 0
    if len(nums) == 1:
        return nums[0]
    if len(nums) == 2:
        return max(nums[0], nums[1])
    # rob houses 0 to n-2 i.e, excluding last house
    result1 = noob_rob(nums[:-1])
    # rob houses 1 to n-1 i.e, excluding first house
    result2 = noob_rob(nums[1:])
    return max(result1, result2)

print(professional_rob([1,2,3]))
```

# Inventory optimization (like Diablo II)

Finally, you got it! You found the Infinite Treasure of the Leprechaun-Dragon! 

Now, you have to decide what to loot. Your inventory is a 5x4 rectangle (5 tiles width, 4 tiles height).

Each item of the treasure also fits in a rectangle and has a value:

Potion of Potionentiality, 1x1, 30 gold,
Jeweled Dog Draught Excluder, 3x1, 150 gold,
Spartan Shield, 2x2, 300 gold,
Palindromic Sword o’Drows, 1x3, 120 gold,
Unobsidian Armor, 2x3, 540 gold,
Wardrobe of Infinite Lions, 20x10, 1337 gold.
You can not turn the items. For example, the Unobsidian Armor is 2 tiles in width and 3 in height. It can not be 3 tiles in width and 2 in height.

Since it’s an Infinite Treasure, there are as many items of each type as you want.

Which items do you take, and how do you organize them in your inventory to have the most significant value in gold?



```python
def optimize_inventory(inventory_width, inventory_height, items,rotated=False):
    # a 3D dp table: [width][height][value, items]
    dp = [[[0, []] for _ in range(inventory_height + 1)] for _ in range(inventory_width + 1)]
    
    # iterate through all possible subspaces of the inventory
    for w in range(1, inventory_width + 1):
        for h in range(1, inventory_height + 1):
            # place each item in the current subspace
            for item_name, (item_width, item_height, item_value) in items.items():
                if item_width <= w and item_height <= h:
                    # value if we place this item
                    curr_value = dp[w - item_width][h][0]
                    total_value = item_value + curr_value
                    
                    # if better value, update the DP table
                    if total_value > dp[w][h][0]:
                        dp[w][h] = [total_value, dp[w - item_width][h][1] + [(item_name, w - item_width, h - item_height)]]
                
                # if rotating the item gives a better result (if dimensions are different)
                if rotated:
                    if item_width != item_height and item_width <= h and item_height <= w:
                        remaining_value = dp[w - item_height][h][0]
                        total_value = item_value + remaining_value
                        
                        if total_value > dp[w][h][0]:
                            dp[w][h] = [total_value, dp[w - item_height][h][1] + [(item_name + " (rotated)", w - item_height, h - item_width)]]
    print(dp)
    return dp[inventory_width][inventory_height]

# Define the inventory and items
inventory_width, inventory_height = 5, 4
items = {
    "Potion of Potionentiality": (1, 1, 30),
    "Jeweled Dog Draught Excluder": (3, 1, 150),
    "Spartan Shield": (2, 2, 300),
    "Palindromic Sword o'Drows": (1, 3, 120),
    "Unobsidian Armor": (2, 3, 540),
    "Wardrobe of Infinite Lions": (20, 10, 1337)
}

# optimize the inventory - no rotated elements
total_value, item_placement = optimize_inventory(inventory_width, inventory_height, items)
print(f"Total value: {total_value} gold")
print("Items to take:")
for item, x, y in item_placement:
    print(f"- {item} at position ({x}, {y})")

# with rotation allowed
total_value, item_placement = optimize_inventory(inventory_width, inventory_height, items, rotated=True)
print(f"Total value: {total_value} gold")
print("Items to take:")
for item, x, y in item_placement:
    print(f"- {item} at position ({x}, {y})")
```
